```python
import pandas as pd
import altair as alt
```

```python
df= pd.read_csv("BankChurners.csv")
```

```python
df.shape
```

```python
features = [
    "Attrition_Flag",
    "Customer_Age",
    "Gender",
    "Marital_Status",
    "Income_Category",
    "Total_Trans_Amt",
    "Credit_Limit",
    "Months_Inactive_12_mon"
]
```

```python
df= df[features].sample(n=5000, ignore_index=True)
```

```python
df.head()
```

```python
(alt.Chart(df).
 mark_bar().
 encode(x="Income_Category", y="mean(Credit_Limit):Q").
 properties(height=300, width=400))
```

```python
df.groupby(["Income_Category"]).agg(Mean_Credit_Limit= ("Credit_Limit", "mean"))
```

```python
(alt.Chart(df).
mark_bar(width=50).
encode(x="Marital_Status", y="mean(Credit_Limit):Q").
transform_filter(
alt.FieldOneOfPredicate(
field="Marital_Status",
oneOf=["Single", "Married", "Divorced"])).
properties(height=300, width=500))
```

```python
sel=alt.selection_multi(
    fields=["Attrition_Flag"], bind= "legend"
)
alt.Chart(df).mark_point(size=50).encode(
x= "Credit_Limit",
y= "Total_Trans_Amt",
color = "Attrition_Flag",
opacity= alt.condition(sel, alt.value(1), alt.value(0.1))
).properties(
height=350, width = 500
).add_selection(sel
               )
```
